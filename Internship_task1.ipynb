```python
# Class4

# Initial list
my_list = [2, 3, 4, 5, 6, 7, 8]
print("Initial List:", my_list)

# Add one element
new_item = int(input("Enter element to add: "))
my_list.append(new_item)
print("Updated List after adding:", my_list)

# Remove an element
remove_item = int(input("Enter element to remove: "))
if remove_item in my_list:
    my_list.remove(remove_item)
    print("Updated List after removing:", my_list)
else:
    print("Element not found, cannot remove")

# Search an element
search = int(input("Enter element to search: "))
if search in my_list:
    print("You found the element")
else:
    print("Not found")

# Add multiple elements
length = int(input("Enter length (how many elements to add): "))
for i in range(length):
    new = int(input("Enter element: "))
    my_list.append(new)

print("Updated List after adding multiple elements:", my_list)


# Chatbot list manager

# Initial empty list
list = []

print("Welcome to the game")

while True:
    choice = input("Type - add/remove/replace/exit/viewlist: ").lower()

    # Add items
    if choice == "add":
        length = int(input("Enter length: "))
        for i in range(length):
            new = int(input("Enter: "))
            list.append(new)
        print("Updated List:", list)

    # Remove item
    elif choice == "remove":
        remove_item = int(input("Enter item: "))
        if remove_item in list:
            list.remove(remove_item)
            print("Updated List:", list)
        else:
            print("Item not found")

    # Replace item
    elif choice == "replace":
        old_value = int(input("Enter old value: "))
        if old_value in list:
            new_value = int(input("Enter new value: "))
            index_old_value = list.index(old_value)
            list[index_old_value] = new_value
            print("Updated List:", list)
        else:
            print("Old value not found")

    # View list
    elif choice == "viewlist":
        print("Current List:", list)

    # Exit
    elif choice == "exit":
        print("Exiting the game ")
        break

    # Invalid choice
    else:
        print("Invalid choice, try again")
```
